Calling the required document loader and creating the indexes

```python
from langchain.document_loaders import UnstructuredPDFLoader
from langchain.indexes import VectorstoreIndexCreator
```

We're using OpenAI's model and embeddings, which will require their API key at a nominal cost 

```python
import os
os.environ["OPENAI_API_KEY"]="YOUR KEY HERE"
```

Now let's mount the drive where we've stored our pdf docs. Here it a Google Drive, but any local drive will work too.

```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
root_dir = "/content/drive/My Drive/"
```

```python
#Identifying the path on the frive to the folder holding your pdf docs
pdf_folder_path = f'{root_dir}/Path to your pdf files on your Google drive/'
```

Next we use LangChain's loader desinged for unstructured pdf files to load each file. This code will create a separate loader object for each file in the folder

```python
loaders = [UnstructuredPDFLoader(os.path.join(pdf_folder_path, fn)) for fn in os.listdir(pdf_folder_path)]
```

Indexing the loaders into the vector db. This will take a while to create a vector for each file 

```python
#I had to load unstructured[local-inference] here before the vector indexes statement would run error free???
!pip install unstructured[local-inference]
index = VectorstoreIndexCreator().from_loaders(loaders)
```

Now you can query your pdf files in a ChatBot-like environment. It can't display an image of a table or a diagram, but it can read the information from the graphics. Use 'index.query_with_sources' to see where in the document it's answer came from.

```python
index.query_with_sources('What is the inductance formula for air-wound coils?')
```

```python
index.query_with_sources('What is the total resistance in a circuit with three 220 ohm resistors in series?')
```
